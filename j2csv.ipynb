```python
import json
import pandas as pd

def json_to_csv_pandas(json_file, csv_file):
    # 读取JSON文件
    with open(json_file, 'r', encoding='utf-8') as f:
        data = json.load(f)
    
    # 提取梦境数据
    dreams = data['dreams']
    
    # 转换为DataFrame
    df = pd.DataFrame(dreams)
    
    # 保存为CSV
    df.to_csv(csv_file, index=False, encoding='utf-8')
    print(f"成功转换 {len(dreams)} 个梦境到 {csv_file}")
    
json_to_csv_pandas('dreams.json', 'dreams.csv')
```
